```python

bins = [18, 25, 35, 45, 55, 65, 75]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '66-75']
sleep_df['Age Group'] = pd.cut(sleep_df['Age'], bins=bins, labels=labels)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age Group', y='Sleep Duration', data=sleep_df, palette='coolwarm')
plt.xlabel('Age Group')
plt.ylabel('Sleep Duration')
plt.title('Distribution of sleep hours by age ')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=sleep_df, x='Sleep Disorder', y='Age', palette='coolwarm')
plt.title('Distribution of Age Among Sleep Disorder Categories')
plt.xlabel('Sleep Disorder')
plt.ylabel('Age')
plt.show()
```

__The older you get, the Increase the Duretion of sleep.__

__The older you get, the more sleep disorders increase.__

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=sleep_df, x='Occupation', y='Sleep Duration', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Sleep Duration Across Different Occupations')
plt.xlabel('Occupation')
plt.ylabel('Sleep Duration (Hours)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=sleep_df, x='Occupation', y='Quality of Sleep', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Average Quality of Sleep by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Quality of Sleep (Higher is Better)')
plt.show()
```

```python
sorted_df = sleep_df.groupby('Occupation')['Sleep Duration'].mean().reset_index().sort_values(by='Sleep Duration')
plt.figure(figsize=(12, 6))
sns.barplot(data=sorted_df, x='Occupation', y='Sleep Duration', palette='coolwarm', estimator=np.mean)
plt.xticks(rotation=45)
plt.title('Average Sleep Duration by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Sleep Duration (Hours)')
plt.show()
```

```python
sorted_df = sleep_df.groupby('Occupation')['Stress Level'].mean().reset_index().sort_values(by='Stress Level')
plt.figure(figsize=(12, 6))
sns.barplot(data=sorted_df, x='Occupation', y='Stress Level', palette='coolwarm', estimator=np.mean)
plt.xticks(rotation=45)
plt.title('Average Stress Level by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Stress Level (Higher means More Stress)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=sleep_df, x='Occupation', hue='BMI Category')
plt.xticks(rotation=45)
plt.title('BMI Category Distribution Across Occupations')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.legend(title='BMI Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=sleep_df, x='Occupation', hue='Sleep Disorder', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Sleep Disorders Distribution Across Occupations')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.legend(title='Sleep Disorder')
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.boxplot(x=sleep_df["Sleep Duration"], y=sleep_df["Quality of Sleep"], palette="coolwarm")
plt.title("Quality of Sleep vs Sleep Duration")
plt.xlabel("Sleep Duration (hours)")
plt.xticks(rotation=45)
plt.ylabel("Quality of Sleep")
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=sleep_df["Sleep Duration"], y=sleep_df["Stress Level"], estimator=np.mean, ci=None, palette="coolwarm")
plt.title("Average Stress Level by Sleep Duration")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Average Stress Level")
plt.show()

plt.figure(figsize=(8, 5))
correlation_matrix = sleep_df[["Sleep Duration", "Stress Level", "Quality of Sleep", "Physical Activity Level"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

__Sleep duration and sleep quality reduce stress.__

__Sleep duration improves sleep quality.__

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=sleep_df["Physical Activity Level"], y=sleep_df["Quality of Sleep"], ci=None, palette="coolwarm")
plt.title("Average Quality of Sleep by Physical Activity Level")
plt.xlabel("Physical Activity Level")
plt.ylabel("Average Quality of Sleep (score)")
plt.show()
```

__There is an improvement in sleep quality with an increase in physical activity level.__
